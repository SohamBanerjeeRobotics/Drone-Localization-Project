```python
# Consider the room to be of dimension 10m * 10m * 3m and resolution is taken to be 0.01 m.
Ambient_Temp = 35 # in degree celsius.
Delta_T_Max = 200 # in degree celsius.
Saturation_time = 300 # in seconds.
Attenuation_coeff = 0.2 # in m^-1
source_pos_x = 0
source_pos_y = 0
source_pos_z = 0
L_T = 2 # in meters.
simulation_time = 1500 # in seconds.
```

```python
import numpy as np
X_Coordinate_time = np.array(np.arange(-5, 5.1, 0.1))
X_Coordinate_time = np.tile(X_Coordinate_time, (101, 1))
X_Coordinate_time = np.tile(X_Coordinate_time, (31, 1, 1))
X_Coordinate_time = np.tile(X_Coordinate_time, (101, 1, 1, 1))

Y_Coordinate_time = np.array(np.arange(-5, 5.1, 0.1)).reshape(101, 1)
Y_Coordinate_time = np.tile(Y_Coordinate_time, (1, 101))
Y_Coordinate_time = np.tile(Y_Coordinate_time, (31, 1, 1))
Y_Coordinate_time = np.tile(Y_Coordinate_time, (101, 1, 1, 1))

Z_Coordinate_time = np.zeros((101, 101))
Z_Coordinate_time = np.tile(Z_Coordinate_time, (31, 1, 1))
for i in range(0,31):
    Z_Coordinate_time[i, :, :] = i/10
Z_Coordinate_time = np.tile(Z_Coordinate_time, (101, 1, 1, 1))

Time_array = np.zeros((31, 101, 101))
Time_array = np.tile(Time_array, (101, 1, 1, 1))
for i in range(0,101):
    Time_array[i, :, :, :] = i + 250

H_Hold = Delta_T_Max * (np.log(1 + (Time_array))/np.log(1 + Saturation_time)) * np.exp((-1) * Attenuation_coeff * np.sqrt(((X_Coordinate_time ** 2) + (Y_Coordinate_time ** 2))))
V_Hold = 5.386 * (((Z_Coordinate_time / L_T) + 0.558) ** 3.205) * np.exp((-4.057) * (Z_Coordinate_time / L_T))
array_temp_dist = Ambient_Temp + (H_Hold * V_Hold) 
# Source Temperature is found at array_temp_dist(501, 501, 151)
```
